```python
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"


# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)

#school_data["budget"] = school_data["budget"].map("${:.2f}".format)
school_data.head()
Total_Budget = school_data["budget"].astype(float).sum()
school_data["Budget per Student"] = school_data["budget"]/school_data["size"]

Total_Budgets = "${:.2f}".format(Total_Budget)
Total_Budgets
school_data
```

```python
student_data_to_load = "Resources/students_complete.csv"
student_data = pd.read_csv(student_data_to_load)
student_data["Avg Score"] = (student_data["reading_score"]+student_data["math_score"])/2
student_data.head()
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="right", on=["school_name", "school_name"])
school_data_complete.head()

```

```python
#Find Total students and schools represented.
Total_schools = school_data_complete["school_name"].unique()
Total_schools
Total_schools.size #Total Schools involved
Total_students = school_data_complete["Student ID"].count() 
Total_students
#Average math and reading scores for all students.
Avg_math = school_data_complete["math_score"].mean()
Avg_reading = school_data_complete["reading_score"].mean()
Avg_reading
Pass_rate = (Avg_reading + Avg_math)/2
# Pass and Fail bins for math and reading.
bins = [0, 69.9, 100]
grouplabelsmath = ["Fail", "Pass"]
grouplabelsnumber = [0, 100]
school_data_complete["Math_Pass"] = pd.cut(school_data_complete["math_score"],bins,labels=grouplabelsmath)
school_data_complete["Reading_Pass"] = pd.cut(school_data_complete["reading_score"],bins,labels=grouplabelsmath)
#Percentage pass on math and reading
PercMath = school_data_complete.loc[school_data_complete["Math_Pass"] == "Pass"]
PercRead = school_data_complete.loc[school_data_complete["Reading_Pass"] == "Pass"]
Math = PercMath["Math_Pass"].count()
Read = PercRead["Reading_Pass"].count()
PercMathPass = Math/Total_students *100
PercReadPass = Read/Total_students *100

school_data_complete.head()
```

```python
#District Summary Dataframe
summary_df = pd.DataFrame({"Total Schools": [Total_schools.size],
                           "Total Students" : [Total_students], 
                           "Total Budget" :  [Total_Budgets],
                           "Average Math Score" : [Avg_math],
                           "Average Reading Score" : [Avg_reading],
                           "% Passing Math" : [PercMathPass] ,
                           "% Passing Reading" : [PercReadPass] ,
                           "% Overall Passing Rate" : [Pass_rate]
                           
                           
                          })
summary_df
```

```python
#School Spending bins based on cost
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_data_complete["School Spending"] = pd.cut(school_data_complete["Budget per Student"],spending_bins,labels=group_names)
SchoolSize = school_data_complete.groupby(["School Spending"])
#Grouping by School Spending, performing average math and reading scores, percentages, and overall percentages.
AvgSchoolMaths = SchoolSize["math_score"].mean()
AvgSchoolReads = SchoolSize["reading_score"].mean()
mathschools = (SchoolSize["Math_Pass"].apply(lambda x: (x=="Pass").sum())/SchoolSize["Student ID"].count())*100
readingschools = (SchoolSize["Reading_Pass"].apply(lambda x: (x=="Pass").sum())/SchoolSize["Student ID"].count())*100

AvgRates = (mathschools + readingschools)/2

#Making the school spending dataframe
SchoolSpending = pd.DataFrame({"Spending Sections":[group_names[0],group_names[1],group_names[2],group_names[3]],
                              "Avg Math": [AvgSchoolMaths[0],AvgSchoolMaths[1],AvgSchoolMaths[2],AvgSchoolMaths[3]],    
                               "Avg Reading": [AvgSchoolReads[0],AvgSchoolReads[1],AvgSchoolReads[2],AvgSchoolReads[3]],
                              " % Pass Math":  [mathschools[0],mathschools[1],mathschools[2],mathschools[3]],
                               " % Pass Read": [readingschools[0],readingschools[1],readingschools[2],readingschools[3]],
                              "Overall Pass Rate":  [AvgRates[0],AvgRates[1],AvgRates[2],AvgRates[3]],
                              })
SchoolSpending.set_index('Spending Sections')
```
